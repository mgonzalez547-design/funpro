#Funciones
Estructura de programacion que permite "cortar" un codigo grande en parte pequeñas y funcionales.
sintaxis de una funcion
```
def nombre_de_la_funcion(d,d1,d2,d3,...,den):
  operacion 1
  operacion 2
  funcion1
  estructura de selccion
  return variable
```
donde:
* d,d1,d2: son los parametros de la funcion
* por dentro pueden ir todas las operaciones
* return vuelve a una variablelocal-->global

existen 4 tipos de funciones
1. sin parametro,sin retorno
```

def funcion1():
operacion1
operacion2

```


````python
#Ejemplo:Desarrollar un progrma que de n notas(donde n es mayor que 1), me diga cual es la mayor nota y cual e
notas=int(input("Ingrese la cantidad de notas:"))
nota=[]
for i in  range(notas):
  n=float(input(f"Ingrese la nota{i+1}:"))
  nota.append(n)
print(nota)
````

````python
 nota_max=max(nota)
nota_min=min(nota)
print(f"La nota maxima es {nota_max} y la peorsita es {nota_min}")
````
