# 2) Determinar si un aÃ±o es bisiesto


Un aÃ±o es bisiesto si es divisible entre 4, excepto si es divisible entre 100, a menos que tambiÃ©n sea divisible entre 400. Es decir, un aÃ±o como 1996 es bisiesto, el 1900 no lo es, pero el aÃ±o 2000 sÃ­ lo es

```python
aÃ±o = int(input("Ingrese un aÃ±o: "))

if aÃ±o % 4 != 0:
    print(f"{aÃ±o} no es bisiesto")
elif aÃ±o % 100 != 0:
    print(f"{aÃ±o} es bisiesto")
elif aÃ±o % 400 == 0:
    print(f"{aÃ±o} es bisiesto")
else:
    print(f"{aÃ±o} no es bisiesto")
```

0# 3) Calculadora de Ãndice de Masa Corporal (IMC)

```python
peso = float(input("Ingrese su peso en kg: "))
altura = float(input("Ingrese su altura en metros: "))

imc = peso / (altura ** 2)
print(f"IMC: {imc:.2f}")

if imc < 18.5:
    categoria = "Bajo peso"

elif imc < 25.0:
    categoria = "Peso normal"
elif imc < 30.0:
    categoria = "Sobrepeso"
else:
    categoria = "Obesidad"

print("CategorÃ­a:", categoria)
```

# 4) MenÃº de opciones calculadora simple con match/case

```python
print("MenÃº de operaciones:")
print("1: Suma")
print("2: Resta")
print("3: MultiplicaciÃ³n")
print("4: DivisiÃ³n")
opcion = input("Seleccione una opciÃ³n (1-4): ")

a = float(input("Ingrese el primer nÃºmero: "))
b = float(input("Ingrese el segundo nÃºmero: "))

match opcion:
    case "1":
        resultado = a + b
        op = "Suma"
    case "2":
        resultado = a - b
        op = "Resta"
    case "3":
        resultado = a * b
        op = "MultiplicaciÃ³n"
    case "4":
        if b == 0:
          print("Error: divisiÃ³n por cero.")
        resultado = a / b
        op = "DivisiÃ³n"
    case _:
        print("OpciÃ³n no vÃ¡lida.")

print(f"{op}: {resultado}")
```

# 5) ClasificaciÃ³n de triÃ¡ngulos
