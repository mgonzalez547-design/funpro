#FUNCIONES
Estructuras de prtogramacion que permite "cortar" un codigo grande en partes pequeñas y funcionales

Sintaxis:

```
def nombre_de_la_funcion(d,d1,d2,d3,...,dn):
  operacion1
  operacion2
  estructura de control
  estructura de seleccion
  return variable

```
donde:
* d,d1,d2: son los paramtreos de la funcion.
* Por dentro pueden ir todas la operaciones
* return convierte una variable local en global


Existen 4 tipos de funciones
1. Sin parametro, sin retorno.
```
def funcion():
  operacion1
  operacion2
```
2. Sin parametro, con retorno.
```
def funcion2():
  operacion1
  operacion2
  return dato
```
3. Con parametro, sin retorno.
```
def funcion3():
 operacion1
 operacion2
```
4. Con parametro, con retorno.
```
def funcion4():
  operacion1
  operacion2
  return dato
```
