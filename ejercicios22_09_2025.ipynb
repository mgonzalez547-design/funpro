# 4) MenÃº de opciones calculadora simple con match/case

```python
print("MenÃº de operaciones:")
print("1: Suma")
print("2: Resta")
print("3: MultiplicaciÃ³n")
print("4: DivisiÃ³n")
opcion = input("Seleccione una opciÃ³n (1-4): ")

a = float(input("Ingrese el primer nÃºmero: "))
b = float(input("Ingrese el segundo nÃºmero: "))

match opcion:
    case "1":
        resultado = a + b
        op = "Suma"
    case "2":
        resultado = a - b
        op = "Resta"
    case "3":
        resultado = a * b
        op = "MultiplicaciÃ³n"
    case "4":
        if b == 0:
          print("Error: divisiÃ³n por cero.")
        resultado = a / b
        op = "DivisiÃ³n"
    case _:
        print("OpciÃ³n no vÃ¡lida.")

print(f"{op}: {resultado}")
```

# 5) ClasificaciÃ³n de triÃ¡ngulos

```python
a = float(input("Lado a: "))
b = float(input("Lado b: "))
c = float(input("Lado c: "))

if a <= 0 or b <= 0 or c <= 0:
    print("Los lados deben ser positivos.")

# desigualdad triangular
if not (a + b > c and a + c > b and b + c > a):
    print("No forma un triÃ¡ngulo.")

if a == b == c: #primero comprueba este, si no es true pasa a los otros
    print("TriÃ¡ngulo equilÃ¡tero (todos los lados iguales).")
elif a == b or b == c or a == c:
    print("TriÃ¡ngulo isÃ³sceles (dos lados iguales).")
else:
    print("TriÃ¡ngulo escaleno (ningÃºn lado igual).")
```
