```python
def analizar_notas(lista_notas):
    n_notas = len(lista_notas)
    nota_max = max(lista_notas)
    nota_min = min(lista_notas)
    i = 0
    sobre_tres = 0

    while i < n_notas:
        if lista_notas[i] >= 3:
            sobre_tres += 1
        i += 1
    promedio=sum(lista_notas)/n_notas
    print("Nota máxima:", nota_max)
    print("Nota mínima:", nota_min)
    print("Cantidad de notas sobre 3:", sobre_tres)

    return nota_max, nota_min, sobre_tres, promedio

notas = [4,3.5,4.5] #dentro del parentesis escriba la Nota 1, Nota 2, etc
analizar_notas(notas)
```
